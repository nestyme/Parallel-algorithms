### Часть 2. Построение и анализ графиков (cashes)

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
plt.style.use('fivethirtyeight')
```

```python
time_differenses = np.zeros(22)
i = -1;
for line in open('output_cashe (2)'):
    if line.strip():           
        i += 1
        try:
            n = float(line) 
            print(n)
            time_differenses[i]=n
        except ValueError:
            pass
print(i)
```

```python
n = [int(x) for x in range(22)]
```

```python
plt.figure(figsize=(18, 6))
print(len(n), len(time_differenses))
plt.plot(n, time_differenses, color='teal')
plt.title("Time Differenses")
plt.xlabel('Iteration')
plt.ylabel("Time")
plt.show()
```
